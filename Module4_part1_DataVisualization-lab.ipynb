## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

# URL of the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_url)

# Check the column names to find the correct column for compensation
print("\nColumn names in the dataset:")
print(df.columns)

# Plot a histogram of the 'ConvertedComp' column
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

# Plot a box plot of the 'Age' column
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.boxplot(df['Age'].dropna())
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# Extract the column with the databases respondents wish to learn next year
# Assuming the column is named 'DatabaseDesireNextYear'
databases = df['DatabaseDesireNextYear'].dropna().str.split(',', expand=True).stack()

# Get the top 5 databases
top_databases = databases.value_counts().head(5)

# Plot a pie chart of the top 5 databases
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(5)))
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Filter data for the age group 30 to 35
age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate medians for 'WorkWeekHrs' and 'CodeRevHrs'
median_work_week_hrs = age_group['WorkWeekHrs'].median()
median_code_rev_hrs = age_group['CodeRevHrs'].median()

# Data for plotting
categories = ['WorkWeekHrs', 'CodeRevHrs']
medians = [median_work_week_hrs, median_code_rev_hrs]

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, medians, color=['skyblue', 'salmon'], edgecolor='k')

# Add a label for the stacked portions
for i, value in enumerate(medians):
    plt.text(i, value/2, f'{value:.1f}', ha='center', va='center', color='white', fontsize=12)

plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Metrics')
plt.ylabel('Median Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Filter data for ages between 45 and 60
filtered_data = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate median ConvertedComp for each age
median_comp_by_age = filtered_data.groupby('Age')['ConvertedComp'].median()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='b')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.xticks(range(45, 61, 1))  # Set x-ticks to show each age in the range
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Count occurrences of each value in the 'MainBranch' column
mainbranch_counts = df['MainBranch'].value_counts()

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Distribution of MainBranch')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
